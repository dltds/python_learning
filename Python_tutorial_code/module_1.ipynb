```python
'''
String are iterable, and we can iterate over
i.e. a thing that has parts that can be accessed one at a time
note: counting/indexing in python starts at 0
'''

num_string = "123456789"

# sub_string = num_string[1]

# start is INCLUSIVE
# stop is EXCLUSIVE
# [start : stop: step]

# we can start counting from the right using a minus

sub_string = num_string[2::5]

sub_string2 = num_string[::5]

sub_string3 = num_string[5::]

sub_string4 = num_string[-1::-1]

print(sub_string)

print(sub_string2)

print(sub_string3)

print(sub_string4)
```
