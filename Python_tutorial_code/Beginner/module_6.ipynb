```python
import pandas as pd

url = r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/1980%202023%20average%20house%20prices.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
df['Period_dt'] = pd.to_datetime(df['Period'], format="%Y-%m", errors='coerce')

# map takes the first argument and applies the same to remaining arguments
df.columns = map(str.lower, df.columns)

# or to specify individual column names we could have used a list
# df.colums = ["column1", "column2"]

df.info()
```

```python
# ploty graph objects - much more powerful

# ploty express - very quick for basic visuals
import plotly.express as px

price_over_time = px.scatter(df, x='period_dt', y='average price all property types')

price_over_time.show()
```

```python
import statsmodels

price_over_time_ols = px.scatter(df,
                                 x='period_dt',
                                 y='average price all property types',
                                 title="Change in average property price (Monlthy/£)",
                                 labels={'period':'Time Period',
                                         "average price all property types":"House Prices"},
                                 trendline="ols")

price_over_time_ols.show()
```

```python
filepath = r'https://raw.githubusercontent.com/data-to-insight/D2I-Jupyter-Notebook-Tools/main/ml-data%20science%20tutorials/data/CINdetails.csv'

cin = pd.read_csv(filepath)

# to get rid of columns we don't want
# data frames have two axes
# 0: rows
# 1: columns
# default of drop is to get rid of rows so we need to set axis=1 for columns
cin.drop('Unnamed: 0', axis=1, inplace=True)

cin.info()
```

```python
import plotly.graph_objects as go

primary_need_bar = px.histogram(cin, x='PrimaryNeedCode')

# need graph_objects to use update_xaxes
primary_need_bar.update_xaxes(categoryorder='category ascending')

primary_need_bar.show()
```

```python
bar_2 = px.histogram(cin, x='PrimaryNeedCode', color='ReferralSource')

bar_2.show()
```

```python
cin['CINreferralDate_dt'] = pd.to_datetime(cin['CINreferralDate'], format="%d/%m/%Y", errors='coerce')
cin['DateOfInitialCPC_dt'] = pd.to_datetime(cin['DateOfInitialCPC'], dayfirst=True, errors='coerce')

# calculates days between 2 dates (as object)
cin['CPCTimeliness'] = (cin['DateOfInitialCPC_dt'] - cin['CINreferralDate_dt'])

# to caculate by days an integer we can use Timedelta from pandas
cin['CPCTimeliness'] = cin['CPCTimeliness']/pd.Timedelta(1, 'd')

timeliness_hist = px.histogram(cin,
                               x='CPCTimeliness',
                               nbins=20,
                               title='Time between Referral and ICPC')

timeliness_hist.show()
```

```python
timeliness_box = px.box(cin,
                        y='CPCTimeliness',
                        color='PrimaryNeedCode')

timeliness_box.show()
```

```python
marginal_scatter = px.scatter(cin,
                              x='CINreferralDate_dt',
                              y='CPCTimeliness',
                              #trendline='ols'
                              marginal_y='violin',
                              #marginal_x='rug',
                              #color='PrimaryNeedCode'
                              )

marginal_scatter.show()
```

```python
facet_plot = px.histogram(cin,
                          x='ReferralSource',
                          facet_col='PrimaryNeedCode',
                          facet_row='ReferralNFA',
                          )

facet_plot.show()
```

```python
CINdetails_no_nans = cin[cin['CPCTimeliness'].notna()]
size_scatter = px.scatter(CINdetails_no_nans, x='CINreferralDate', y='PrimaryNeedCode', size='CPCTimeliness')
size_scatter.show()
```
