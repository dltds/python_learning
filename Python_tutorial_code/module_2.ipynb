```python
# Functions and Methods
# print(argument) <- an example of a function
# variable_of_type.method() <- method 

# List methods include
# .append() - adds iterables as a single element
# .extend() - adds iterables as individual elements e.g. add a list as individual elements
# .remove()
# .sort()

more_colours = ["red", "blue", "green", "beige", "red2"]
print(more_colours)

more_colours.append("brown") # mutates
print(more_colours)

more_colours.extend(["amber","pink"]) # 
print(more_colours)

more_colours.remove("red2") # removes one element with a given value
print(more_colours)
```

```python
# Dictionaries store variables or other things in key : value pairs
# key : value
# keys must be immutatable not mutatable -> keys must be either a string or a number

fav_colours = {"Will":"Black",
               "Clare":"Yellow",
               "Marcel":{"Monday":["Blue","Yellow"],
                         "Rest of Week":"Grey"}
               }

type(fav_colours)

print(fav_colours)

# Dictionaries always stay in the same order
#  You access disctionaries via Keys

print(fav_colours["Marcel"])
print(fav_colours["Marcel"]["Monday"]) # accessing nested elements
print(fav_colours["Marcel"]["Monday"][0]) # accessing nested elements

print(fav_colours.keys())
type(fav_colours.keys())
list(fav_colours.keys())

print(fav_colours.values())

print(fav_colours.items())
list(fav_colours.items()) # returns tuples
```
